# Network Graph Library Improvments

## Overview

To impove the fidelity of interactive network graphs here are some changes that I made to the library `nx_altair`. 
I incorperated better arrows, better scaled axis, better labels and more methods of interacting with the graph

```python
import altair as alt
import pandas as pd
import numpy as np
import networkx as nx
import modified_nx_altair.nx_altair as nxa
import original_nx_altair as nxa_o
import altair as alt
import pandas as pd
import bs4
import numpy as np
from os import path
from altair import datum
import time
import altair_saver
```

This is what the original ouptut of the graphs looked like.

```python
# Add weights to nodes and edges
G = nx.gn_graph(n=20)

# Compute positions for viz.
pos = nx.spring_layout(G)

for n in G.nodes():
    G.nodes[n]['weight'] = np.random.randn()

for e in G.edges():
    G.edges[e]['weight'] = np.random.uniform(1, 10)

for e in G.edges():
    G.edges[e]['label'] = np.random.uniform(1, 10)


# Draw the graph using Altair
viz = nxa_o.draw_networkx(
    G, pos=pos,
    node_color='weight',
    cmap='viridis',
    width='weight',
    edge_color='black',
)

# Show it as an interactive plot!
viz.interactive()
```

In the original there are scaling issues and arrows and labels aren't supported.

```python
viz = nxa.draw_networkx(
        G=G,
        pos=pos,
        node_label='name:N',
        width='width:Q',
        #     cmap='accent',
        node_tooltip=[alt.Tooltip('name:N', title='Alias'),
                        alt.Tooltip('day:N', title='Day Joined'),
                        alt.Tooltip('reviews_given:Q', title='Reviews Given in Month'),
                        alt.Tooltip('incoming:Q', title='Reviews Recieved in Month'),
                        alt.Tooltip('kind:N', title='User Type'),
                        alt.Tooltip('chapter_total:Q', title='Chapters Writen During in Month in Fandom'),
                        'betweenness', 'degree']
)

vizz.interactive()
```
